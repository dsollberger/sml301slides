```python
# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=False,
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")

# Create a pipeline
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=500,
    do_sample=False,
)
```

# Prompt Engineering

## Ingredients of a Prompt

```python
# Prompt components
persona_staffer = "You are a Congressional staffer and college graduate. You have been asked to provide a historical perspective of the issue.\n"
persona_lobbyist = "You are a business lobbyist who seeks to discourage further government regulation in artificial intelligence.  You seek to build upon anti-regulation sentiment.\n"
persona_ngo = "You are a volunteer at a non-governmental organization (NGO) that raises awareness about data privacy and the corpuses that are used in the training of large language models.\n"
persona_legal = "You are a paralegal at a law firm that has been tasked with explaining the legal subtleties in the current American laws about artificial intelligence.\n"

instruction = "Provide possible debate points and support a stance.\n"
context = "The list should express several key points that can help a debate team and quickly state important information for the scene.\n"
data_format = "Create a bullet-point overview that summarizes the debate stance and also provide a concise paragraph that initiates the debate.\n"
audience = "We are at a USA government meeting with a subcommitee of Congress. We are hoping to inform our Congress members about the issues about governmental regulation in artificial intelligence.\n"
tone = "The tone should be informative and professional.\n"
text = wiki_text  # the underlying data
data = f"Text to summarize: {text}"
```

## Congressional Staffer

```python
# build prompt
query = persona_staffer + instruction + context + data_format + audience + tone + data
# print(query)
messages = [
    {"role": "user", "content": query}
]
# print(tokenizer.apply_chat_template(messages, tokenize=False))
```

```python
# Generate the output
outputs = pipe(messages)
print(outputs[0]["generated_text"])
```
